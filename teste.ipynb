```python
import librosa
import librosa.display
import numpy as np
from deep_audio import Visualization, Audio
import matplotlib.pyplot as plt
import scipy
```

```python
path2 = 'archive/VCTK-Corpus/VCTK-Corpus/wav48/p225/p225_006.wav'

signal, rate = Audio.read(path2, sr=24000, normalize=True)
```

```python
mfcc = librosa.feature.mfcc(signal, rate, n_mfcc=20, n_fft=2048, hop_length=512, lifter=22)
Visualization.plot_cepstrals(mfcc.T, title='Mel-Frequency Ceptrum Coefficients', x_label='Índices de Frame', y_label='MFCC Índices', show=True, cmap='viridis', fig_name='images/apresentacao/mfcc')
```

```python
Visualization.plot_spectrogram(signal, rate=rate, title='Densidade Espectral', x_label='Tempo (s)', y_label='Frequência (Hz)', show=True, close=True, cmap='viridis', fig_name='images/apresentacao/power')

```

```python
Visualization.plot_audio(signal, rate, title='Onde de Som', y_label='Amplitude', x_label='Tempo (s)', show=True, fig_name='images/apresentacao/wave.png')
```

```python
lpc = librosa.lpc(signal, 240)
Visualization.plot(lpc, title='Linear Predictive Coding', y_label='Coeficientes de erro', x_label='Ordem', show=True, fig_name='images/apresentacao/lpc')
```

```python
signal = Audio.unnormalize(signal)
lpcc = Audio.lpcc(signal, rate)
Visualization.plot_cepstrals(lpcc, title='Linear Predictive Coding Coefficients', x_label='Frame Indexes', y_label='LPCC Indexes', show=True, cmap='viridis', fig_name='images/apresentacao/lpcc.png')
```
